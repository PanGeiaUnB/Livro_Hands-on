## 3. Introdução teórica


### 3.1 Problema

O Sistema Móvel Celular vem crescendo a cada ano, ganhando mais usuários e se tornando mais importante na vida dos indivíduos. Esses fatores culminaram no surgimento de novos desafios a serem superados nas próximas gerações dessas tecnologias, como no 5G.

Dentre esses desafios, é preciso encontrar maneiras de otimizar o uso das frequências disponíveis, já que a quantidade de usuários vem crescendo de maneira considerável. 

Nesse contexto, concebeu-se a ideia de criar um tipo de comunicação entre dispositivos que não fosse mediada pela estação Rádio-Base, denominada *Device to Device* (D2D). Esse tipo de comunicação se mostrou muito promissor para o contexto do 5G, de forma que já se vislumbram diversas aplicações para a mesma. Para um melhor entendimento, leia essa [breve introdução](http://www2.decom.ufop.br/imobilis/d2d-em-redes-5g/) em que o autor detalha o que se espera do D2D no contexto do 5G.

### 3.2 Sistema Móvel

Em um sistema de comunicações móveis convencional, cada comunicação é alocada em uma faixa de frequência diferente das faixas ocupadas pelas demais frequências da célula. Isso faz com que uma comunicação não cause interferência nas demais, já que, por conta das características da modulação OFDM, uma faixa de frequência é ortogonal às outras. Nesses sistemas, cada uma dessas faixas de frequência é conhecida como *Resource Block* (RB).

Cada canal tem um número finito de *Resource Blocks*, de forma que o número de usuários atendidos simultaneamente pelo sistema também é limitado. Uma solução para aumentar essa capacidade é inserir mais de um dispositivo por *Resource Block* e encontrar maneiras de garantir que a interferência entre os dispositivos não seja alta ao ponto de inviabilizar as comunicações prioritárias.

Antes de propormos maneiras de mitigar os efeitos da interferência, devemos entender como a interferência ocorre nesse tipo de sistema.

#### 3.2.1 Interferência entre comunicações

Como explicado, comunicações alocadas na mesma faixa de frequência causam interferência entre si. O nível de interferência de um sinal é medido por uma grandeza muito similar à SNR, apresentada no Experimento 1, a Relação Sinal-Ruído mais Interferência (SNIR - *Signal-to-Noise plus Interference Ratio*). Essa variável pode ser equacionada como:

$$
SNIR = \frac{P}{N+I}
$$

em que $P$ é a potência do sinal recebido, $N$ é a potência do ruído contido no sinal e $I$ é a potência dos sinais interferentes.

Como visto nos experimentos de detecção de símbolos em canais AWGN, a taxa de erro na detecção de símbolos (SER) é inversamente proporcional à SNR. Estendendo o problema para canais em que há interferência, a SER se torna inversamente proporcional à SNIR. Dessa forma, a definição da SNIR de uma comunicação indica qual é a porcentagem de símbolos que são corretamente detectados no dispositivo receptor. 

Como consequência disso, surge o **Teorema de Shannon–Hartley**, que define que a taxa de transmissão máxima $C$ de um canal de banda $B$ é calculada por:

$$ C = B \cdot log_2(SNIR+1) $$

Uma breve introdução sobre o teorema pode ser encontrada [aqui](https://pt.wikipedia.org/wiki/Teorema_de_Shannon%E2%80%93Hartley).

Para garantir que uma determinada comunicação seja efetuada com sucesso, é necessário garantir que a taxa de transmissão esteja acima de um certo limiar ou, analogamente, garantir que a SNIR dela esteja acima de um determinado patamar, que é chamado de **SNIR mínima** da comunicação. Assim, maximizar a SNIR das comunicações de um RB significa maximizar a taxa de transmissão, o que é muito desejado para o sistema.

Entendido como funcionam as interferências em um sistema de comunicações, precisamos compreender como o ambiente afeta um sinal transmitido por um dispositivo em uma determinada posição até chegar em outro dispositivo que vai receber esse sinal, completando a comunicação.

#### 3.2.2 Perda no caminho

Como já foi dito em outros experimentos, as comunicações do sistema móvel utilizam o ar como canal de transmissão. Sendo assim, dentro desse canal existem muitos fatores que enfraquecem a potência do sinal que chega ao receptor, entre eles podemos citar a topologia do terreno, as construções urbanas e a flora local.

Todos esses fatores servem de anteparo para o sinal enviado, refletindo-o, refratando-o e absorvendo-o. Assim, quando um sinal é enviado através do ar, uma parte da potência se perde, e parte da potência que se perde chamamos de **perda no caminho**, ou *path loss*.

Essa perda depende tanto de fatores estruturais da célula quanto de configurações do sistema de comunicações. Entre elas, podemos citar as principais:
*   Distância entre os dispositivos;
*   Frequência do sinal;
*   Altura das antenas; e
*   Características do ambiente.

A imagem abaixo ilustra esse fenômeno:

image.png

Ao longo dos anos, surgiram vários modelos que se propuseram a modelar a perda no caminho, levando em conta os fatores citados acima. A maioria desses modelos surgiu de forma empírica, procurando uma equação que se ajustasse às diferentes configurações que uma célula poderia ter. Vários deles foram validados e são vastamente utilizados em simulações.

Neste experimento, utilizaremos o modelo ITU-1411 para células de ambiente urbanos. A equação que define a perda nesse modelo é:

$$ L = 21.2\log_{10}d + 29.2 + 21.1\log_{10}fc $$

em que $L$ é a perda do caminho em dB, $d$ é a distância entre os dispositivos transmissor e receptor em metros e $fc$ é a frequência do sinal, em MHz.


#### 3.2.3 Modelagem da interferência

Entendendo como os sinais são transmitidos no sistema móvel, só precisamos entender como uma comunicação causa interferência nas demais.

O receptor de uma comunicação recebe, além do sinal transmitido pelo transmissor dessa comunicação, todos os outros sinais enviados por todos os dispositivos que estejam transmitindo na mesma frequência. Assim, todos esses sinais são transmitidos através do ar, sofrendo perdas de caminho até chegar a um receptor, determinando uma certa SNIR. 

Como a frequência é constante, a perda no caminho de uma comunicação só depende da distância entre transmissor e receptor, o que é válido tanto para a comunicação desejável quanto para as interferentes. 

Dessa maneira, as SNIRs das comunicações se tornam função apenas da potência de cada sinal transmitido e da distância entre todos os dispositivos receptores e transmissores da célula.

## 4. Aprendizado por reforço

O aprendizado por reforço (RL - *Reinforcement Learning*) é um campo de Inteligência Artificial que tenta, através do recebimento de recompensas, ensinar um agente inteligente a decidir ações adequadas para uma determinada situação vivenciada. De maneira intuitiva, o processo é semelhante ao adestramento de um cachorro que aprende a realizar algumas tarefas a partir do recebimento de biscoitos.

#### 4.1.1 Ambiente

O ambiente é a representação do problema que será utilizada para permitir que um algoritmo de *Deep Reinforcement Learning* aprenda a tomar decisões baseadas no estado em que o ambiente se encontra. Esse elemento é responsável por modelar a transição de estados, ou seja, como a ação realizada modifica o ambiente como um todo. 

Para o exemplo que adotamos, o ambiente é o jogo como um todo, os objetos, espaços, inimigos e toda modelagem.

#### 4.1.2 Agente

O agente é um algoritmo inteligente capaz de compreender o estado em que o ambiente se encontra e, a partir disso, decidir quais ações executadas gerarão recompensas altas. 

Assim, o papel do agente é aprimorar a sua política de decisão de ações baseada no estado, visando sempre encontrar a política que lhe traga o máximo retorno possível, sendo este a soma das recompensas recebidas. A política do agente é denotada como $\mu$.

No exemplo adotado, o agente faria o papel que o ser humano faz ao jogar o *game*, decidindo quais ações devem ser executadas pelo personagem.

image.png

#### 4.1.3 Estado

O estado, denotado como $s$, é definido como a situação do ambiente em determinado instante, ou seja, como o ambiente se encontra no momento da observação feita. A representação do estado faz parte da modelagem do ambiente, já que um único problema apresenta diferentes maneiras de modelagem do vetor que representa o estado, de forma que é necessário decidir qual dessas maneiras é a mais adequada para o entendimento do agente.

Como citado, não existe apenas uma modelagem do estado, mas podemos defini-lo no problema como a imagem que aparece na tela do monitor, já que é isso que o nosso cérebro utiliza para interpretar o ambiente e fazer as decisões.

image.png

#### 4.1.4 Recompensa

A recompensa é a variável responsável por avaliar o quão positiva ou negativa foi uma determinada ação executada. A partir dessa grandeza é possível ensinar a rede a decidir as ações que serão tomadas, visando sempre maximizar a recompensa recebida.

Essa variável, denotada por $r$, é, geralmente, um número real, cujo valor é definido pelas chamadas funções de recompensa, que são funções responsáveis por calcular a recompensa da ação executada a partir de padrões definidos na modelagem do ambiente.

Apresentados os conceitos básicos de RL, precisamos entender o que é o *Deep Reinforcement Learning*.

### 4.2 Deep Reinforcement Learning

O DRL é uma consequência de toda teoria de RL desenvolvida ao longo de anos. Em 2013, [Volodymyr Mnih propõe utilizar uma rede neural profunda como agente de um problema de RL](https://www.cs.toronto.edu/~vmnih/docs/dqn.pdf), que até então tentava modelar os agentes utilizando, principalmente, modelos estatísticos. A aplicação do artigo, inclusive, era produzir um agente capaz de jogar *games* clássicos do console Atari.

Dessa forma, o DRL surge da ideia de se incorporar o *Deep Learning* em problemas de *Reinforcement Learning*, visando resolver problemas de complexidade elevada. A partir dessa união, precisamos entender como a rede neural aprenderá a tomar decisões a partir de um estado do ambiente. 

#### 4.2.1 Redefinição do problema

A partir das últimas seções, nós já entendemos que o agente aprenderá a decidir as ações a partir das recompensas que ele obtiver, de maneira que recompensas altas recebidas após uma ação encorajará o agente a realizar ações parecidas quando estiverem em estados semelhantes. Caso contrário, o agente entenderá que é preciso evitar determinadas ações em determinados estados.

Entretanto, problemas de DRL geralmente não são tão simples. Veja a figura abaixo.

image.png

Supondo que o objetivo principal do agente do *game* seja completar a fase do jogo, mas que a coleta de moedas também lhe traga um valor adicional de recompensa, qual seria a ação que traria maior recompensa para o personagem?

Analisando o estado, provavelmente essa ação seria andar para a frente e coletar a moeda. Entretanto, em poucos instantes após tomar essa ação, o personagem cairia no buraco e reiniciaria a fase, recebendo uma recompensa bastante negativa.

Com esse exemplo, percebemos claramente que não precisamos de um agente que seja capaz de maximizar a recompensa de cada ação tomada, nós precisamos de um algoritmo capaz de **maximizar o retorno $R_t$**. 

Antes de definir matematicamente o retorno precisamos entender dois conceitos:

*   **Evento:** é cada instante de tempo em que o agente lê  a entrada e toma uma ação;
*   **Episódio:** é o conjunto de eventos que ocorrem em uma existência do agente, seja ela longa ou curta, exitosa ou falha. No nosso exemplo, um episódio seriam todos os eventos desde o início da fase até o reinício, seja porque o personagem morreu, seja porque a fase foi completada.

Entendido esses dois conceitos, podemos definir o retorno:

$$ R_t = \sum_{k=t}^{T} r_k $$

em que $t$ é o evento do episódio, $T$ é o número de eventos contidos em um episódio e $r_k$ é a recompensa recebida no evento $k$, que por sua vez é o índice que vai de $t$ até o final do episódio $T$. Repare que o retorno é denotado levando em consideração o evento $t$, já que o somatório que o define depende do evento em que este é calculado, de forma que calcular o retorno no primeiro evento ($R_0$) é diferente de calculá-lo em um evento $t$ futuro ($R_t$).

Entendido o papel do agente de maximizar o retorno, precisamos pensar em como ele aprenderá isso. 

A partir do exemplo mostrado, fica claro que precisamos de um agente que consiga prever situações futuras, isto é, voltando a última imagem, precisamos de um agente que compreenda que apesar de não ser a ação que maximiza a recompensa, saltar o buraco vai trazer ao agente a possibilidade de receber mais recompensas no futuro, maximizando o retorno. 

Pense que, ao tomar essa ação, o personagem vai para outro estado, de onde ele pode traçar diferentes trajetórias e não uma única, como mostrado na figura abaixo.

image.png

Sendo assim, para que o agente consiga decidir quais ações devem ser tomadas, ele precisa, estando em um determinado estado, conseguir identificar a ação que o levará para um estado em que a média de retorno de todas as trajetórias possíveis é máxima. Essa média de retornos possíveis define uma variável chamada **Valor do estado-ação $Q(s,a)$**, que é matematicamente expressa como:

$$ Q(s,a) = \mathbb{E}_{\mu}[R_t|s_t=s, a_t=a] $$

em que $\mu$ é a política que o agente usa para definir as ações a serem tomadas e $\mathbb{E}$ é o cálculo da esperança dos retornos de todas as possíveis trajetórias a partir do momento que o agente toma a ação $a$ no estado $s$.

Entretanto, pense no número de possibilidades que se abrem a partir do momento que o personagem pula o buraco. O número de combinações de ações possíveis é enorme, de forma que é impossível que o algoritmo simule cada uma dessas possibilidades em tempo hábil para calcular $Q(s,a)$ com precisão e, então, definir qual ação deve ser tomada.

É nesse momento que nasce a ideia de se introduzir uma rede neural que consiga **estimar** a variável $Q(s,a)$ com precisão razoável, de forma que seja possível, estando em um estado $s$, estimar a variável $Q(s,a)$ para as diferentes ações possíveis, e assim, escolher a ação cuja estimativa de retorno seja máxima.


#### 4.2.2 Deep Q-Learning - DQL

A *Deep Q-Network* foi a primeira arquitetura de DRL, sendo precursora de quase todas as outras que surgiram posteriormente. Ela funciona utilizando exatamente o que foi explicado na seção anterior. A imagem abaixo ilustra o seu funcionamento.

image.png

Podemos ver na imagem que o agente é uma DNN que observa o estado $s$ do ambiente e calcula o valor de $Q$ para as $N$ ações possíveis. Trazendo para o exemplo tratado, as $N$ ações seriam as possibilidades que o personagem tem, como correr, pular, ir para frente ou para trás e etc..

A partir do que foi exposto acima, já seria possível elaborar um algoritmo que conseguisse decidir quais ações devem ser tomadas pelo personagem do jogo. 

Entretanto, essa rede apresenta uma clara limitação: ela só consegue resolver problemas em que as ações possíveis são **discretas**. Isso é perceptível ao perceber que a rede retorna apenas **qual** ação deve ser tomada, e não **como** ela deve ser executada. Passemos a um outro problema para que isso fique mais claro.

Pense em um jogo de corrida, em que o agente seja um piloto responsável por acelerar e guiar um carro, como exemplificado abaixo.

image.png

Repare que ambas as ações são **contínuas**, isto é, não basta o agente saber que precisa acelerar e virar o volante para alguma direção, ele precisa entender **o quanto** ele precisa acelerar e **o quanto** ele precisa rotacionar o volante.

Voltemos, então, ao DQL. Para implementar uma DQN, neste caso, precisaríamos discretizar todas as posições do volante, bem como todos os níveis possíveis do acelerador, considerando que cada uma dessas discretizações é uma ação com um $Q(s,a)$ calculável. 

Perceba que essa não é uma abordagem eficaz mesmo para um problema relativamente simples como esse. Identificando essa limitação, desenvolveu-se outra arquitetura que consegue ultrapassar essa barreira.

#### 4.2.3 Actor-Critic

É importante ressaltar que o termo *Actor-critic* **não** designa uma arquitetura de DRL, e sim uma característica que várias arquiteturas da área incorporaram, entre elas o A2C, A3C e o DDPG. Nesse notebook, apresentaremos e utilizaremos o DDPG.

Dizer que uma arquitetura é *actor-critic*, significa dizer que ela utiliza **duas** redes neurais no seu processo de aprendizagem: uma que toma a ação e outra que critica a ação tomada, daí surge o nome *actor-critic*.

A rede do ator recebe o estado como entrada e produz uma saída, podendo ser um número ou um vetor. No exemplo do *game* de corrida, o ator poderia receber a imagem da tela do monitor como entrada e produzir um vetor de tamanho 2, em que:

*   O primeiro elemento fosse um número entre -1 e 1, que indicaria o quanto o volante deveria ser rotacionado para um lado ou para o outro;
*   O segundo fosse um valor de 0 a 1, que indicaria o quanto o acelerador deveria ser pressionado.

Após a tomada de decisão do ator, o crítico vai estimar o valor de $Q(s,a)$ a partir do estado e dessa ação realizada, tentando avaliar se a ação tomada foi boa ou ruim.

Assim, percebemos que essas arquiteturas desagregam o tomador de decisão e o estimador do $Q(s,a)$, de forma que a saída do ator pode ser de qualquer natureza, cabendo ao crítico realizar a estimativa do retorno a partir da ação tomada. Entendido isso, precisamos entender como as duas redes aprendem.

#### 4.2.4 Deep Deterministic Gradient Policy - DDPG

Repare que o problema para o agente ainda é o mesmo: precisamos desenvolver um algoritmo capaz de maximizar o retorno. Nesse cenário, como treinar o ator?

Como o $Q$ é uma estimativa do retorno recebido, é possível treinar o ator através da saída do crítico: se a saída deste for alta, o ator entenderá que agiu de maneira adequada, caso contrário, o ator terá que mudar a sua política de decisão de ação, modificando os seus parâmetros.

Certo, mas como treinaremos o crítico para que ele consiga fazer estimativas de $Q(s,a)$ de maneira acurada, de forma que o ator possa "confiar" nessa estimativa para atualizar os seus pesos?

#### 4.2.5 Equação de Bellman

Para compreender o aprendizado do crítico, voltaremos à definição da variável $Q(s,a)$. Repare que utilizando as propriedades de um somatório, podemos escrever o retorno como:

$$ R_t = r_t + \sum_{k=t+1}^{T} r_k \\
   R_t = r_t + R_{t+1}
 $$

Nos próximos passos, não entraremos na matemática da dedução por acreditar que o seu entendimento demanda um pouco mais de tempo, mas deixamos uma [referência](https://lilianweng.github.io/lil-log/2018/02/19/a-long-peek-into-reinforcement-learning.html) que faz essa dedução completa.

Para nós, o importante é compreendermos a noção de que o retorno presente pode ser decomposto na recompensa atual adicionada ao retorno do próximo evento e que isso pode ser estendido para o valor do estado-ação, por ser uma variável em função do retorno. 

$$ Q(s,a) = r_t + Q(s',a') $$

em que $s'$ é o estado futuro (no evento $t+1$) e $a'$ é a ação tomada no evento $t+1$. Entretanto, em um cenário em que tanto a transição de estados, quanto a tomada de decisão não são conhecidas, é necessário tratarmos com as **esperanças** dessas variáveis. Assim:

$$ Q(s,a) = \mathbb{E}_{s'\mathtt{\sim}P}\{r_t + \mathbb{E}_{a'\mathtt{\sim}\mu}[Q(s',a')]\} $$

em que $P$ é a probabilidade de transição dos estados e $\mu$ é a política para tomada de decisão adotada pelo ator. Essa última equação é conhecida como **Equação de Bellman**, e é a base para o aprendizado de quase todas as arquiteturas de DRL. 

Na maioria das referências, veremos um termo $\gamma$ multiplicando a esperança de $Q(s',a')$, chamado de fator de desconto. Essa variável serve para o algoritmo dar mais importância às experiências recentes, em detrimento das que estão em um futuro mais distante. Nessa explicação, ele foi suprimido por simplificação.

Vejamos, agora, como podemos utilizar essa equação para ensinar a rede neural do crítico.


#### 4.2.6 Processo de aprendizagem

Pensemos no início do treinamento do algoritmo: a rede neural do crítico faz uma estimativa para o valor de $Q$, mas por ainda não estar treinada, essa estimativa não é confiável. Nesse sentido, o que precisaríamos fazer para a rede aprender a melhorar essa estimativa? Qual seria o valor esperado que utilizaríamos para calcular o custo que seria usado no otimizador da rede?

Voltemos à Equação de Bellman. Repare que, se a nossa forma de calcular o $Q$ é a partir de uma rede neural, precisaríamos usá-la para calcular ambos os lados da equação:

*    No primeiro, passaríamos à rede o estado e ação atuais ($s$ e $a$) como entrada e receberíamos a estimativa de $Q(s,a)$ como saída;
*    No segundo, passaríamos à rede o estado e ação futuras ($s'$ e $a'$) como entrada e receberíamos a estimativa de $Q(s',a')$ como saída.

Para completar a equação, precisaríamos, ao executar a ação $a$ no estado $s$, coletar $r_t$, que é calculada de maneira determinística pelo próprio ambiente e enviada para o algoritmo.

Apesar de a equação de Bellman ser matematicamente bem definida, de forma que os dois lados da equação precisem ser iguais, ao utilizar uma rede neural para o cálculo do $Q$, estamos realizando apenas uma estimativa para essa variável, de forma que, na prática, haverá divergência entre os dois lados da equação.

Mas como podemos usar tudo isso para ensinar uma rede totalmente leiga a aprender a estimar o valor de $Q$? 

Pense que os dois lados da equação de Bellman, calculados a partir da utilização da estimativa da rede neural, possuem grandes incertezas, principalmente no início do processo de aprendizagem. Entretanto, como foi dito, $r_t$ é uma variável coletada diretamente do ambiente, **não havendo incerteza alguma em seu valor**.

Assim, graças a esse fato, podemos dizer que, utilizando redes neurais para estimar o valor de $Q$, o cálculo do lado direito da equação de Bellman possui **menos incerteza** do que o lado esquerdo da equação, que depende apenas da saída da rede.

Dessa forma, é possível treinar a rede neural do crítico utilizando uma função de perda que calcule o custo considerando $Q(s,a)$ como a estimativa feita pela rede e o lado direito da equação como o valor esperado.

Perceba que o valor esperado para treinar a rede depende da saída da própria rede, algo que é inesperado e contraintuitivo, mas pouco a pouco os valores de recompensa recebidos vão direcionando a rede para a convergência, de forma que as estimativas vão melhorando a cada época de treinamento.

O fato apontado acima é o principal responsável pelas redes de DRL serem mais instáveis e possuírem convergência mais lenta do que as redes de ML/DL tradicionais, o que faz com que essas arquiteturas precisem ter um treinamento mais longo, além de usarem diversas técnicas para ajudar na convergência do aprendizado.

A imagem abaixo ilustra o treinamento de um DDPG em um único evento. Na figura, as setas verdes significam a passagem de entradas e saídas até a estimativa dos dois lados da equação de Bellman e as setas vermelhas mostram o processo de atualização dos parâmetros das redes a partir dos custos calculados.

image.png

Na imagem acima, a função de custo do crítico ilustrada pode ser de diferentes naturezas. A mais clássica para identificar a diferença entre duas grandezas é o Erro Quadrático Médio (*Mean Squared Error* - MSE) e é ela que será utilizada. 

Para atualizar o ator, nós queremos promover o **aumento** do $Q(s',a')$ a cada ação executada, estimulando ações com alto valor do estado-ação. Pensando então no processo de aprendizado, sabemos que a maioria dos algoritmos se baseiam na **diminuição** do gradiente da perda, como adequar o problema?

Uma solução é utilizar a perda como $-Q(s',a')$, de forma que, quanto maior o $Q$, menor a função de perda $-Q$ e, consequetemente, menor serão as atualizações dos parâmetros da ator. Em um caso contrário, a rede entenderá que as suas ações estão distantes do objetivo e que a política de tomada das ações deve ser modificada.

Para entender mais sobre o processo de treinamento do DDPG, o pseudo-código pode ser encontrado neste [link](https://towardsdatascience.com/deep-deterministic-policy-gradients-explained-2d94655a9b7b).



#### 4.2.7 Estratégias de treinamento do DDPG

Como foi dito, o treinamento das arquiteturas de DRL é um processo instável e, por isso, adotam-se algumas técnicas para tentar diminuir essa instabilidade. Essa seção se propõe a apresentar brevemente algumas dessas técnicas, mesmo que de forma superficial, encorajando que o aluno pesquise mais a fundo futuramente.

##### 4.2.7.1 Memória para treinamento

A despeito do processo mostrado acima, percebeu-se que atualizar os pesos do ator e do crítico a cada evento causava instabilidade no processo de treinamento.

Percebendo isso, alguns autores propuseram a utilização de uma memória que armazenasse as experiências vividas pelo algoritmo, de onde seria escolhido aleatoriamente um conjunto de experiências para atualizar os parâmetros das redes (*batch*).

Perceba que o treinamento em *batchs* utiliza diferentes experiências ao mesmo tempo. Esse processo aproxima o treinamento mostrado na última figura à equação de Bellman, levando em consideração as esperanças contidas no cálculo da equação.

Esse [artigo](https://arxiv.org/pdf/1710.06574.pdf) discute os efeitos da utilização dessa memória de maneira mais profunda, para aqueles que se interessarem.

##### 4.2.7.2 Mecanismo de exploração

Muitos algoritmos de DRL conseguem encontrar uma boa política de tomada de decisão, estimulando o ator a continuar tomando ações parecidas, o que ajuda na convergência da rede.

Entretanto, essa pode não ser a **política ótima**, de forma que há como melhorar o desempenho da rede, mas, pelo fato de receber recompensas relativamente altas, a rede não modifica seus parâmetros, continuando estagnada em uma política medíocre.

Uma maneira de se evitar isso é utilizando um mecanismo de exploração. Basicamente, a ideia por trás dessa técnica é, durante o processo de treinamento, introduzir aleatoriedade na tomada de decisão, forçando o agente a tomar uma decisão que não foi planejada pela rede neural.

Em problemas com ações discretas, é possível fazer isso escolhendo algumas ações de maneira aleatória. Já em problemas com ações contínuas, o mais comum é que se adicione ruído na saída da rede neural, variando essas ações de maneira aleatória, como mostrado nessa [publicação](https://towardsdatascience.com/deep-deterministic-policy-gradients-explained-2d94655a9b7b), na seção "*Exploration*".

Entretanto, recentemente surgiu uma nova abordagem para essa exploração. Propôs-se que, em vez de adicionar o mecanismo de exploração após o funcionamento da rede, seria possível adicionar esse mecanismo **durante** a atuação da rede neural.

A ideia é que nós adicionemos um ruído aleatório nos parâmetros da rede, alterando todos os pesos desta. Essa aleatoriedade garante que a rede tome ações diferentes mesmo em situações parecidas, trazendo a ideia de exploração. Essa [tese de mestrado](https://matthiasplappert.com/publications/2017_Plappert_Master-thesis.pdf) abordou o assunto de forma bastante simples e completa. Será essa a estratégia utilizada no nosso código. 

##### 4.2.7.3 Utilização de redes auxiliares

Os autores perceberam que atualizar completamente as redes do ator e do crítico a cada *batch* de experiências traria grandes mudanças na politica do agente a cada evento, o que poderia causar instabilidade no processo de aprendizagem.

Uma proposta para solucionar isso foi utilizar redes auxiliares, denominadas redes *targets*, que são redes atualizadas a cada evento, mas de maneira suavizada, evitando mudanças bruscas nos seus parâmetros. Isso faz com que as estimativas dessa rede se tornem mais "confiáveis" e mais estáveis.

Sabendo disso, o processo de aprendizagem se utiliza dessas redes para calcular o lado direito da equação de Bellman, diminuindo a instabilidade do processo de treinamento do algoritmo.

Essa [publicação](https://towardsdatascience.com/deep-deterministic-policy-gradients-explained-2d94655a9b7b) também trata das redes *target*.

Depois dessa introdução sobre DRL, *actor-critic* e DDPG, organizaremos o problema.

## 5.Organizando o problema

Em um *Resource Block* de uma célula do Sistema Móvel, consideraremos o problema havendo:
- $1$ dispositivo móvel $M$ se movendo pela célula.
- $N$ dispositivos D2D, indiciados com $k$ de $1$ a $N$, se movendo ao longo da célula. 

Pensando na construção de um ambiente para a inserção de um algortimo de Reinforcement Learning, podemos definir que:

**Estado:** posição do dispositivo móvel, posição de cada D2D, SNIR da comunicação do dispositivo móvel e SNIR da comunicação de cada D2D. Tudo isso será sintetizado numa tupla:

$$ state = [(x_M,y_M), (x_{k=1},y_{k=1}), ... , (x_{k=N},y_{k=N}), \\ SNIR_M, SNIR_{k=1}, ... , SNIR_{k=M}] \\ $$

**Ações:** Potência do dispositivo móvel e dos dispositivos D2D:

$$action = [P_M, P_{k=1}, ... , P_{k=N}] \\ $$

**Objetivo e recompensa:** Encontrar os maiores $SNIR_k$ possíveis, diminuir ao máximo a sua variância (evitar "injustiças" na comunicação) e não permitir que a $SNIR_M < L$. Assim:

$$ r_1 = \dfrac{\gamma_1}{N} \sum_{k=1}^{N}SNIR_k \\ $$
$$ r_2 = \gamma_2 \sum_{k=1}^{N}Pot_k \\ $$
$$ r_3 = \left\{\begin{array}{ll}
\gamma_3 * (2-2^{-SNIR_M + (SNIR_{min} + 1)}), \:\:\:\: SNIR_M \geq SNIR_{min} \\  
-2\gamma_3 + \gamma_3^{\frac{SNIR_M}{5}},\:\:\:\: cc \\
\end{array} \right. \\ $$

para $\gamma_1, \gamma_2$ e $\gamma_3$ sendo os fatores de escala.

Por fim:

$$ R = \frac{r_1 + r_2 + r_3}{2} $$





Como $r3$ possui uma definição mais complexa, plotamos o eu gráfico abaixo em função da SNIR do MUE e considerando 10 dB como SNIR mínima da comunicação prioritária.

image.png

Assim, percebemos que $r3$ penaliza o agente de maneira linear à diferença entre a SNIR do MUE e a SNIR mínima, se aquela é menor do que esta. 

Para o caso de essa SNIR estar acima da SNIR mínima, $r3$ aumenta rapidamente, até encontrar um limitante superior (próximo a 15 dB), de forma que aumentar essa SNIR acima da SNIR mínima não trará um grande aumento de recompensa. Esse comportamento estimula o agente a manter a SNIR do MUE acima da SNIR mínima, mas não recompensa o agente por continuar aumentando essa SNIR indefinidamente. 

## 6. Implementação

Para os alunos interessados, recomendamos que abram o código e tentem entender o que faz cada método do ambiente, mas isso não é fundamental para a sequência do experimento.

Sendo assim, focaremos na elaboração do DDPG em si.

Iniciaremos com a implementação da memória, que é composta por 2 métodos principais:

*   `push`: responsável por adicionar elementos na memória;
*   `sample`: responsável por selecionar um *batch* de experiências aleatoriamente.

```python
Transition = namedtuple('Transition',
                        ('state', 'action', 'next_state', 'reward'))

class ReplayMemory(object):
    def __init__(self, capacity):
        self.capacity = capacity
        self.memory = []
        self.position = 0

    def push(self, *args):
        """Saves a transition."""
        if len(self.memory) < self.capacity:
            self.memory.append(None)
        self.memory[self.position] = Transition(*args)
        self.position = (self.position + 1) % self.capacity

    def sample(self, batch_size):
        return random.sample(self.memory, batch_size)

    def __len__(self):
        return len(self.memory)
```

Com a definição da memória, podemos passar para a definição das redes neurais que serão utilizadas.

O intuito do experimento é desenvolver um algoritmo inteligente capaz de definir as potências de transmissão de cada dispositivo transmissor em um *Resource Block*, de maneira a maximizar a SNIR de cada comunicação, **sempre** evitando violar a SNIR mínima da comunicação prioritária.

Sendo assim, precisamos definir o que passaremos como entrada para que a rede entenda como deve definir as potências. Para esse experimento, definiu-se que seriam passadas as coordenadas geográficas de todos os dispositivos da célula (para o entendimento da interferência entre comunicações) e as SNIRs de todas as comunicações (para o entendimento de como está a situação imediatamente antes de redefinir as potências das comunicações). Esses dados serão colocados lado a lado em um vetor unidimensional simples que será utilizado como entrada das redes.

A arquitetura utilizada tanto para o ator quanto para o crítico é uma DNN, similar à utilizada no Experimento 1. Existe uma pequena diferença entre elas, já que a rede do crítico recebe não só o estado como entrada, mas também a ação definida pelo ator, para que assim ela julgue se uma determinada ação em um determinado estado foi uma boa escolha.

Ambas as arquiteturas estão expostas na célula abaixo. Recomendamos que vocês pesquisem o que faz as funções `nn.LayerNorm` e `weight.data.mul_`, pois isso será questionado ao final do experimento. 

```python
class cnn(nn.Module):
    def __init__(self, input_size, outputs):
        super(cnn, self).__init__()
        self.fcl = nn.Linear(input_size, 32)
```

```python
class Actor(nn.Module):
    def __init__(self, input_size, outputs):
        super(Actor, self).__init__()
        self.linear1 = nn.Linear(input_size, 32)
        self.ln1 = nn.LayerNorm(32)

        self.linear2 = nn.Linear(32, 64)
        self.ln2 = nn.LayerNorm(64)

        self.linear3 = nn.Linear(64, outputs)
        self.linear3.weight.data.mul_(0.1)
        self.linear3.bias.data.mul_(0.1)

    # Called with either one element to determine next action, or a batch
    # during optimization. Returns tensor([[left0exp,right0exp]...]).
    def forward(self, x):
        out = nn.functional.relu(self.ln1(self.linear1(x)))
        out = nn.functional.relu(self.ln2(self.linear2(out)))
        out = self.linear3(out)
        return out

class Critic(nn.Module):
    def __init__(self, state_size, action_size, outputs):
        super(Critic, self).__init__()
        self.linear1 = nn.Linear(state_size, 32)
        self.ln1 = nn.LayerNorm(32)

        self.linear2 = nn.Linear(32 + action_size, 64)
        self.ln2 = nn.LayerNorm(64)

        self.linear3 = nn.Linear(64, outputs)
        self.linear3.weight.data.mul_(0.1)
        self.linear3.bias.data.mul_(0.1)

    # Called with either one element to determine next action, or a batch
    # during optimization. Returns tensor([[left0exp,right0exp]...]).
    def forward(self, state, action):
        s1 = nn.functional.relu(self.ln1(self.linear1(state)))
        out = torch.cat((s1,action), dim=1)
        out = nn.functional.relu(self.ln2(self.linear2(out)))
        out = self.linear3(out)
        return out
```

Agora, definiremos o treinador do algoritmo. Basicamente, o treinador será quem organizará a atuação da rede.

Inicializaremos o treinador, configurando o tamanho do *batch*, os parâmetros de aprendizagem, os parâmetros do método de exploração (vide seção 4.2.7.2), as redes do ator e crítico e a memória que será utilizada, além de algumas variáveis que serão utilizadas para plotagem.

```python
class Treinador():
    def __init__(self,amb):
        self.BATCH_SIZE = 64 # tamanho do batch
        self.GAMMA = 0.9 # fator de desconto
        self.tau = 0.05 # fator de suavização das atualizações

        # Configuração do ruído nos parâmetros
        self.sigma_0 = 0.12
        self.alfa = 1.01
        self.delta = 1e-1
        self.T_adapt = 5

        # Inicialiazação do ator e do crítico
        self.amb = amb

        self.n_actions = len(self.amb.communications)
        self.input_size = len(self.amb.state)

        self.actor_net = Actor(self.input_size, self.n_actions).to(device)
        self.actor_noise_net = Actor(self.input_size, self.n_actions).to(device)
        self.actor_target_net = Actor(self.input_size, self.n_actions).to(device)
        
        self.critic_net = Critic(self.input_size, self.n_actions, 1).to(device)
        self.critic_target_net = Critic(self.input_size, self.n_actions, 1).to(device)
  
        self.actor_target_net.load_state_dict(self.actor_net.state_dict())
        self.actor_target_net.eval()
        self.actor_noise_net.load_state_dict(self.actor_net.state_dict())
        self.actor_noise_net.eval()
        self.critic_target_net.load_state_dict(self.critic_net.state_dict())
        self.critic_target_net.eval()

        # Parâmetros de aprendizagem
        self.optimizer_critic = torch.optim.AdamW(self.critic_net.parameters(), lr=0.001)
        self.optimizer_actor = torch.optim.AdamW(self.actor_net.parameters(), lr=0.0001)

        self.loss_d = nn.MSELoss()
        self.loss_critic = nn.MSELoss()

        # Instância da memória
        self.memory = ReplayMemory(10000)

        # Variáveis de plotagem
        self.actor_loss = None
        self.critic_loss = None
        self.actor_losses = []
        self.critic_losses = []

        self.recompensas_teste = []
        self.SNIRs_teste = []
        self.potencias_teste = []

        self.recompensas = []
        self.SNIRs = []
        self.potencias = []
```

A partir de agora, desenvolveremos funções que serão colocadas dentro da classe `Treinador` como métodos, isto é, vamos criar essas funções separadamente por uma questão de organização, mas elas pertencerão à classe criada acima.

Sabendo que queremos desenvolver um agente, precisamos definir a função que vai selecionar uma ação  a cada evento de um episódio, o que é feito na função `select_action`, disposta na célula abaixo. Repare que essa função já introduz o ruído nos parâmetros como método de exploração.

```python
def select_action(self, state, param_noise=None):
    self.actor_net.eval()
    if param_noise is not None: 
        mu = self.actor_noise_net(state.float())
    else:
        mu = self.actor_net(state.float())

    self.actor_net.train()
    mu = mu.data
      
    return mu
```

```python
#Adição da função na classe Treinador
Treinador.select_action = select_action
```

A função `load_model` é uma função auxiliar para o caso de nós termos salvado algum modelo treinado de um agente e de um crítico e quisermos utilizá-los.

```python
def load_model(self, actor_model, critic_model):
    self.actor_net.load_state_dict(torch.load(actor_model, map_location=torch.device(device)))
    self.critic_net.load_state_dict(torch.load(critic_model, map_location=torch.device(device)))
```

```python
Treinador.load_model = load_model
```

A função `weightSync` é responsável por sincronizar as redes de maneira suavizada, evitando grandes mudanças bruscas nas redes *target*. A ideia por traz dela também se encontra nessa [publicação que já foi citada](https://towardsdatascience.com/deep-deterministic-policy-gradients-explained-2d94655a9b7b), na seção *Target Network Updates*.

```python
def weightSync(self, target_model, source_model):
    for parameter_target, parameter_source in zip(target_model.parameters(), source_model.parameters()):
        parameter_target.data.copy_((1 - self.tau) * parameter_target.data + self.tau * parameter_source.data)
```

```python
Treinador.weightSync = weightSync
```

Já a função `optimize_model` é responsável por otimizar os dois modelos, aplicando a equação de Bellman para definir os valores de perda do ator e do crítico.

Apesar de ter algumas linhas iniciais para adequar os dados que são utilizados, a função não é complexa, de forma que recomendamos fortemente que o aluno se esforce para entender como a equação de Bellman é aplicada e como as redes são atualizadas. Deixaremos uma marca no código, apontando a parte mais importante a ser entendida.

```python
def optimize_model(self, t):
    if len(self.memory) < self.BATCH_SIZE:
        return
    transitions = self.memory.sample(self.BATCH_SIZE)

    batch = Transition(*zip(*transitions))

    non_final_mask = torch.tensor(tuple(map(lambda s: s is not None, batch.next_state)), device=device, dtype=torch.bool)
    non_final_next_states = torch.cat([s.unsqueeze(00) for s in batch.next_state
                                                if s is not None])
    
    state_batch = torch.cat(batch.state)
    action_batch = []
    for tupla in batch.action:
        action_batch.append(np.array(tupla.cpu()))
    self.action_batch = torch.from_numpy(np.array(action_batch)).to(device)
    reward_batch = torch.cat(batch.reward)

    ##### Parte importante ########
    # Grandezas para o cálculo da equação de Bellman
    next_state_values = torch.zeros(self.BATCH_SIZE, device=device)
    next_actor_action_value = self.actor_target_net(non_final_next_states.float())
    with torch.no_grad():
        next_state_values[non_final_mask] = self.critic_target_net(non_final_next_states.float(), next_actor_action_value.float()).squeeze()

    # Equação de Bellman
    expected_state_action_values = (next_state_values * self.GAMMA) + reward_batch

    # Otimização dos modelos
    self.optimizer_critic.zero_grad()
    state_action_values = self.critic_net(state_batch.float(), self.action_batch.float())
    self.critic_loss = self.loss_critic(state_action_values.double(), expected_state_action_values.unsqueeze(1).double())
    
    self.critic_loss.backward()
    self.optimizer_critic.step()

    self.optimizer_actor.zero_grad()            
    gradient = -self.critic_net(state_batch.float(), self.actor_net(state_batch.float()).float())
    self.actor_loss = torch.mean(gradient)
    self.actor_loss.backward()
    self.optimizer_actor.step()

    ##### FIM da parte importante ########
    
    self.weightSync(self.actor_target_net, self.actor_net)
    self.weightSync(self.critic_target_net, self.critic_net)

    return state_batch
```

```python
Treinador.optimize_model = optimize_model
```

Definidos todos esses parâmetros, podemos desenvolver a função `treinar` que, obviamente, vai treinar o algoritmo de fato, combinando todas as ações adicionadas acima.

```python
def treinar(self,num_episodes, actor_model=None, critic_model=None, param_noise = True, plot_results = False):
    self.num_episodes = num_episodes       

    self.violacoes = []
    for i_episode in range(num_episodes):
        # Initialize the environment and state
        total_snirs, total_potencias, total_reward = 0,0,0
        state = self.amb.initial_state().to(device)

        for parameter_target, parameter_source in zip(self.actor_noise_net.parameters(), self.actor_net.parameters()):
            parameter_target.data.copy_(parameter_source.data + self.sigma_0*torch.randn(parameter_source.data.shape).to(device))

        violacao = 0            
        for t in count():
            # Select and perform an action
            action = self.select_action(state, param_noise=param_noise)
            
            self.actor_noise_net.train(),self.actor_net.train(),self.critic_net.train()

            reward, done = self.amb.step(action)
            reward = torch.tensor([reward], device=device)

            self.amb.update(action)
            # Observe new state
            next_state = self.amb.state
            
            state, next_state = state.unsqueeze(0).float().to(device), torch.Tensor(next_state).to(device)
            self.memory.push(state, action, next_state, reward)
            state = next_state

            self.amb.duracoes -= 0.1
            if done:
                break

            state_batch = self.optimize_model(t)

            if self.amb.SNIRs[0] < self.amb.SNIR_min:
                violacao += 1 

            total_potencias +=  np.array(self.amb.pots, dtype='float64')
            total_snirs += np.array(self.amb.SNIRs, dtype='float64')
            total_reward += reward.item()
            
            #if t % 100 == 0:
             #   print('SNIRS: ' , np.array(self.amb.SNIRs), '\n', 'Potencias: ', action.cpu().numpy())

        if param_noise == True:
            if (i_episode+1) % self.T_adapt == 0: 
                unpertubed_action = self.select_action(state_batch)
                perturbed_actions = self.action_batch

                d = self.loss_d(unpertubed_action, perturbed_actions)
                d = torch.mean(d)
                if d <= self.delta:
                    self.sigma_0 = self.sigma_0*self.alfa
                else:
                    self.sigma_0 = self.sigma_0/self.alfa

        print('Epsode ', i_episode, ': ', 'Recompensa=', np.around(total_reward/t,3), '; Potencias=', np.around(total_potencias/t,3))
        self.recompensas.append(round(total_reward/t,4))
        self.SNIRs.append(tuple(np.around(total_snirs/t, decimals=4)))
        self.potencias.append(tuple(np.around(total_potencias/t, decimals=4)))
        self.violacoes.append(violacao/t)

        if i_episode == 0:
            self.plot_interf = self.amb.array_interf
            self.plot_rx = np.array(self.amb.pots_rx)
        else:
            self.plot_interf = np.concatenate((self.plot_interf.reshape(len(self.amb.communications), len(self.amb.communications), -1),self.amb.array_interf.reshape(len(self.amb.communications),len(self.amb.communications),-1)), axis=2)
            self.plot_rx = np.concatenate((self.plot_rx.reshape(len(self.amb.communications),-1), np.array(self.amb.pots_rx).reshape(len(self.amb.communications),-1)), axis=1)
           
        if self.critic_loss != None:
            self.critic_losses.append(self.critic_loss.item())
            self.actor_losses.append(self.actor_loss.item())

        self.amb.reset()

    print('Complete')

    if actor_model != None and critic_model != None:
        torch.save(self.actor_net.state_dict(), actor_model)
        torch.save(self.critic_net.state_dict(), critic_model)
```

```python
Treinador.treinar = treinar
```

Finalizada a função de treino, só precisamos desenvolver uma função de plotagem dos resultados que serão obtidos durante o treinamento, para que vejamos se o resultado é satisfatório. Assim, definimos a função `plot_results`.

```python
def plot_results(self, epsode, SNIRs=[], recompensas=[], potencias=[], violacoes=[], losses=[], interferencia=[], pots_rx=[]):
    epsode += 1
    SNIRs_user, SNIRs_d2d = [], []
    Pot_user, Pot_d2d = [], []
    for i in range(len(SNIRs)):
            SNIRs_user.append(SNIRs[i][0])
            SNIRs_d2d.append(SNIRs[i][1:])
            Pot_user.append(potencias[i][0])
            Pot_d2d.append(potencias[i][1:])

    if SNIRs != []:
        fig1 = plt.figure(1, figsize=(10,7))
        plt.clf()
        ax1 = plt.subplot(311)
        ax1.set_title('Recompensas')
        plt.setp(ax1.get_xticklabels(), visible=False)

        plt.plot(range(1,epsode+1), recompensas, label='Recompensa')
        plt.grid(True) #Mostra a grade


        ax2 = plt.subplot(312)
        ax2.set_title('SNIR')
        plt.setp(ax2.get_xticklabels(), visible=False)

        plt.plot(range(1,epsode+1), SNIRs_user, label='MUE')
        plt.plot(range(1,epsode+1), SNIRs_d2d, label='D2D', linestyle='--')

        plt.grid(True) #Mostra a grade
        plt.legend()


        ax3 = plt.subplot(313)
        ax3.set_title('Potências de entrada')
        plt.setp(ax3.get_xticklabels(), visible=True)

        plt.plot(range(1,epsode+1), Pot_user, label='MUE')
        plt.plot(range(1,epsode+1), Pot_d2d, label='D2D', linestyle='--')

        plt.grid(True) #Mostra a grade
        plt.legend()

    if interferencia != []:
        fig2 = plt.figure(2)
        plt.clf()
        for i in range(interferencia.shape[0]):
            ax = plt.subplot(len(self.amb.communications), 1,i+1)
            ax.set_ylabel('C' + str(i+1))
            if i+1 != len(self.amb.communications):
                plt.setp(ax.get_xticklabels(), visible=False)

            percent_interf = 100*interferencia[i]/interferencia[i].sum(axis=0)
            plt.stackplot(range(1,epsode+1), percent_interf, labels=('C1', 'C2', 'C3', 'C4'))

        plt.suptitle('Perfil da interferência (%)')          
        plt.legend()


    if violacoes != []:
        fig3 = plt.figure(3)
        plt.clf()
        markerline ,_,_ = plt.stem(violacoes)
        plt.setp(markerline, markersize = 0.5)
        plt.title('Probabilidade de outage')
        plt.grid()

    if losses != []:
        fig4 = plt.figure(4)
        plt.clf()
        plt.plot(losses[0], label='Loss do critic')
        plt.plot(losses[1], label='Loss do actor')
        plt.title('Loss do agente e do crítico')
        plt.legend()
        plt.grid()

    if pots_rx != []:
        fig5 = plt.figure(5)
        plt.clf()
        proporcao = 100*np.array(interferencia.sum(axis=1)/(pots_rx + interferencia.sum(axis=1))).reshape(len(self.amb.communications),-1)

        for i in range(proporcao.shape[0]):
            ax = plt.subplot(len(self.amb.communications), 1,i+1)
            ax.set_ylabel('C' + str(i+1))
            if i+1 != len(self.amb.communications):
                plt.setp(ax.get_xticklabels(), visible=False)


            plt.plot(range(1,epsode+1), proporcao[i,:], label='C'+str(i+1))
            
        plt.suptitle('Porcentagem de interferência no sinal recebido (%)')          
        plt.legend()
    plt.show()
```

```python
Treinador.plot_results = plot_results
```
